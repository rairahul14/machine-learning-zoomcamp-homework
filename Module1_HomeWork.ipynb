```python
first_7records=df1[['vehicle_weight','model_year']].head(7)
first_7records
```

```python
X = first_7records.to_numpy()
print(X)
```

```python
XTX = X.T @ X
```

```python
import numpy as np
```

```python
Invert_XTX = np.linalg.inv(XTX)
```

```python
y = np.array([1100, 1300, 800, 900, 1000, 1100, 1200])
y
```

```python
w = Invert_XTX @ X.T @ y 
```

```python
result_sum = np.sum(w)
result_sum
```
